# import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# load the iris dataset

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

# load the iris data from seaborn libraries


```python
data = sns.load_dataset('iris')
data.head()
```

```python
data['species'].value_counts()
```

# checking for missing values

```python
data.isnull().sum()
```

# divide the iris data into X and y

```python
X = data.drop('species', axis=1)
y = data['species']
```

# Apply train test split on x and y

# import train test split from sklearn libraries

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size= 0.25, random_state = 1)
X_train.shape, X_test.shape
```

# Apply Decision Tree from train dataset with the help of RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 100, random_state=1)
rfc.fit(X_train, y_train)
```

# make prediction on the testing dataset

```python
y_pred = rfc.predict(X_test)
```

# Evaluate the accuracy of the model

# import the importent libraries

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(X_train, y_train)
```

# plot the dtc

```python
from sklearn import tree
fig =plt.figure(figsize=(25,20))
_ = tree.plot_tree(dtc,
                   feature_names = X.columns,
                   filled =True)
```
